```python
import numpy as np
import math

base_delta = 3*math.pow(10, -8)
c = 3*math.pow(10, 8)
d = 50
```

```python
# Генерируем N согласно распределению Пуасонна, labmda = 2
l = 2
n_vect = []
for i in range(1000):
  u = np.random.uniform()
  j = 0
  p = math.exp(-l)
  f = p
  while u >= f:
    p *= l/(j+1)  
    f += p
    j += 1
  n_vect.append(j)
print(n_vect)
```

```python
def simulate(n):
  if n == 0:
    return 0

  # места где фотон появился
  dist_arr = [d*(i-0.5) for i in np.random.uniform(size=n+1)]

  # скорость к одному из детекторов
  vspd_arr = [math.sin(2*math.pi*i) for i in np.random.uniform(size=n+1)]

  # calculating the time needed to reach the sensor for each particle
  time_arr = [] # считаем времена достижения одного из детекторов
  for s, v in zip(dist_arr, vspd_arr): # ???
    dist_to_travel = 0
    if v == 0: continue
    if v > 0:
      dist_to_travel = d/2-s
    else:
      dist_to_travel = -d/2-s
    time_arr.append(dist_to_travel/(c*v))

  # подсчет дельт между попаданиями фотонов
  time_arr.sort()
  time_deltas = []
  for first, second in zip(time_arr[:-1], time_arr[1:]):
    time_deltas.append(second-first)

  # проверка, будет ли задетектировано попадание  
  if any(delta < base_delta for delta in time_deltas):
    return 1
  else:
    return 0
```

```python
probs = []
for n in n_vect:
  probs.append(simulate(n))
```

```python
print("Вероятность активации =", probs.count(1)/len(probs))
```
