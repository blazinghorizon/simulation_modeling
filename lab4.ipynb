```python
import numpy as np
import math
```

```python
N = 300   # число экспериментов
n = 20    # число здоровых клеток
m = 100   # общее число клеток
k = 10    # число оставшихся здоровых клеток, которое надо превысить
l = 0.1   # лямбда для вычислений
```

```python
cumul = 0
for i in range(N):
  # через какое время клетки погибнут (показательное распределение)
  cells_lifespan = [-1/l * math.log(1-x) for x in np.random.uniform(size=m)]

  # последняя не умершая по времени клетка
  cells_lifespan.sort(reverse=True)
  prob_k = cells_lifespan[k-1]

  # вероятность, что злокачественная клетка умрет за время, не превышающее найденное значение
  pre_prob = 1-math.exp(-l*prob_k)
  
  # т.к. время жизни клеток независимо -> умножить на количество клеток
  prob = math.pow(pre_prob, n)
  cumul += prob
```

```python
print("Вероятность:", cumul/N)
```
