```python
n = 200  
m = 1000 
k = 100  # ัะธัะปะพ ะพััะฐะฒัะธััั ะทะดะพัะพะฒัั ะบะปะตัะพะบ, ะบะพัะพัะพะต 
gen_lambda = 0.1
```

```python
cumul = 0
for i in range(50):
  # generating ttl for healthy cells
  # ัะตัะตะท ะบะฐะบะพะต ะฒัะตะผั ะพะฝะธ ัะดะพัะฝัั
  healthy_vect = [-1/gen_lambda * math.log(1-x) for x in np.random.uniform(size=m)]
  healthy_vect.sort(reverse=True)
  print("t =", healthy_vect)

  # finding k greatest value
  # ะฟะพัะปะตะดะฝัั ะฝะต ัะผะตััะฐั ะฟะพ ะฒัะตะผะตะฝะธ ะบะปะตัะบะฐ
  t_k = healthy_vect[k-1]
  print("tk =", t_k)

  # ะฒะตัะพััะฝะพััั, ััะพ ะทะปะพะบะฐัะตััะฒะตะฝะฝะฐั ะบะปะตัะบะฐ ัะผัะตั ะทะฐ ะฒัะตะผั, ะฝะต ะฟัะตะฒััะฐััะตะต t_k
  pre_prob = 1-math.exp(-gen_lambda*t_k)
  print(pre_prob)
  # ั.ะบ. ะฒัะตะผั ะถะธะทะฝะธ ะบะปะตัะพะบ ะฝะตะทะฐะฒะธัะธะผะพ -> ัะผะฝะพะถะธัั ะฝะฐ ะบะพะปะธัะตััะฒะพ ะบะปะตัะพะบ
  prob = math.pow(pre_prob, n)
  print("prob =", prob, "\n")
  cumul += prob
```

All what is left to do is to calculate the estimation by averaging the value

```python
print("The estimated probability is:", cumul/50)
```
