```python
import numpy as np
import math

base_x = -10
base_y = 0
base_z = 0
base_r = 9.9
N = 1000
```

========================================

–ì–∞–º–º–∞ == 1

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ (x - x0)^2 + (y - y0)^2 + (z - z0)^2 = r^2

=======================================

‚à´Ô∏Å‚à´Ô∏Åùêª2 Vcos(ùúÉ)dùúÉdœï = 1 ‚üπ V = 1/2pi

P(œï) = [0, pi/2] ‚à´cos(ùúÉ)/2pi dùúÉ = 1/2pi

œï ~ [0, 2pi]

========================================

P(ùúÉ) = [0, 2pi] ‚à´cos(ùúÉ)/2pi dœï = cos(ùúÉ)

F(ùúÉ) = [0, ùúÉ] ‚à´cos(ùúÉ) dùúÉ = sin(ùúÉ)

ùúÉ = arcsin(U)

========================================

–î–ª—è —Ä–∞—Å—á–µ—Ç–∞ V(w) –Ω—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç –ª–∏ –ø—Ä—è–º–∞—è —à–∞—Ä:

x = sin(ùúÉ)cos(œï)t

y = sin(ùúÉ)sin(œï)t

z = cos(ùúÉ)t

========================================

–ü–æ–¥—Å—Ç–∞–≤–∏–º –≤ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –∏ —Ä–∞—Å–∫—Ä–æ–µ–º —Å–∫–æ–±–∫–∏:

(x^2+y^2+z^2)t^2 - 2(xx0+yy0+zz0)t + (x0^2+y0^2+z0^2-R^2) = 0

–ü–æ–ª—É—á–∞–µ–º –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ, –µ—Å–ª–∏ —É –Ω–µ–≥–æ –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –∫–æ—Ä–µ–Ω—å - –∑–Ω–∞—á–∏—Ç –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç, –∏–Ω–∞—á–µ - –Ω–µ—Ç

========================================

```python
def get_x_y_z(theta, phi):
  return (
    math.sin(theta) * math.cos(phi),
    math.sin(theta) * math.sin(phi),
    math.cos(theta)
  )

def simulate_with_w():
  sample_vals = []
  for _ in range(N):
    # –ø–æ–ª—É—á–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤
    phi = np.random.uniform(0, 2*math.pi)
    theta = math.asin(np.random.uniform())

    # –ø–æ–ª—É—á–∞–µ–º –≤–µ–∫—Ç–æ—Ä
    x, y, z = get_x_y_z(theta, phi)

    # —Ä–µ—à–∞–µ–º –∫–≤–∞–¥–∞—Ä—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ
    b = -2 * (x*base_x + y*base_y + z*base_z)
    c = math.pow(base_x, 2) + math.pow(base_y, 2) + math.pow(base_z, 2) - math.pow(base_r, 2)
    d = math.pow(b, 2) - 4*c

    # –µ—Å–ª–∏ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç >= 0, –∑–Ω–∞—á–∏—Ç –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç
    if d >= 0:
      # –≤ —Å–ª—É—á–∞–µ —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –∫–æ—Å–∏–Ω—É—Å—ã –ø–æ–¥ –∏–Ω—Ç–µ–≥—Ä–∞–ª–æ–º —Å–æ–∫—Ä–∞—Ç—è—Ç—Å—è
      sample_vals.append(1)# math.cos(theta))
    else:
      # –µ—Å–ª–∏ –º–µ–Ω—å—à–µ 0, –∑–Ω–∞—á–∏—Ç –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç
      sample_vals.append(0)
  return sample_vals

imp_sampl = simulate_with_w()

print(imp_sampl)
print(np.average(imp_sampl))
```

To generate uniformally distributed vectors on the hemishpere, the angle values ùúÉ and œï should be calculated as follows:

ùúÉ = arccos(1 - U)

œï ~ [0, 2pi]

```python
def simulate_monte_carlo():
  sample_vals = []
  for _ in range(N):
    # –ø–æ–ª—É—á–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤
    phi = np.random.uniform(0, 2*math.pi)
    theta = math.acos(1 - np.random.uniform())

    # –ø–æ–ª—É—á–∞–µ–º –≤–µ–∫—Ç–æ—Ä
    x, y, z = get_x_y_z(theta, phi)

    # —Ä–µ—à–∞–µ–º –∫–≤–∞–¥–∞—Ä—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ
    b = -2 * (x*base_x + y*base_y + z*base_z)
    c = math.pow(base_x, 2) + math.pow(base_y, 2) + math.pow(base_z, 2) - math.pow(base_r, 2)
    d = math.pow(b, 2) - 4*c

    # –µ—Å–ª–∏ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç >= 0, –∑–Ω–∞—á–∏—Ç –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç
    if d >= 0:
      sample_vals.append(math.cos(theta))
    else:
      # –µ—Å–ª–∏ –º–µ–Ω—å—à–µ 0, –∑–Ω–∞—á–∏—Ç –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç
      sample_vals.append(0)
  return sample_vals

std_sampl = simulate_monte_carlo()

print(std_sampl)
print(np.average(std_sampl))
```
