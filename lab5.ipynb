```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import math

n = 20
no_of_expirements = 1000
```

```python
dist_vect = []
wght_vect = []

# Функция получения возможных продолжений пути
def get_available_dirs(cur_x, cur_y, history):
  lambda_list = []
  if [cur_x+1, cur_y] not in history:
    lambda_list.append(lambda x, y: [x+1, y])
  if [cur_x-1, cur_y] not in history:
    lambda_list.append(lambda x, y: [x-1, y])
  if [cur_x, cur_y+1] not in history:
    lambda_list.append(lambda x, y: [x, y+1])
  if [cur_x, cur_y-1] not in history:
    lambda_list.append(lambda x, y: [x, y+1])
  return lambda_list

def process():
  step = 0
  history = [[0, 0]]
  weights = []
  while step < n:
    x, y = history[step]
    directions = get_available_dirs(x, y, history)
    if not directions:
      history.pop()
      weights.pop()
      step -= 1
    else:
      history.append(np.random.choice(directions)(x, y))
      weights.append(len(directions))
      step += 1
  distance = math.sqrt(history[-1][0]**2 + history[-1][1]**2)
  return distance, history, weights

codes = [Path.LINETO for i in range(21)]
codes[0] = Path.MOVETO

for i in range(no_of_expirements):
  d, p, w = process()
  print(f"N = {i+1}, длина пути = {d}")
  dist_vect.append(d)
  wght_vect.append(np.multiply.reduce(w))
```

```python
average_weight = np.average(wght_vect)
eval_vect = []
print(dist_vect)
print(wght_vect)
for d, w in zip(dist_vect, wght_vect):
  eval_vect.append(w/average_weight * d)
eval = np.average(eval_vect)
print(eval)
```
